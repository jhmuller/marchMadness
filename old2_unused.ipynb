```python
import os
import sys
import pandas as pd
import time
import datetime
import yaml
from microprediction import MicroWriter
from getjson import getjson
import matplotlib as mpl
import matplotlib.pyplot as plt
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python

def make2(val):
    return str(val) if val >= 10 else "0"+str(val)

dfs = []
base = datetime.datetime.today()
numdays = 200
date_list = [base - datetime.timedelta(days=x) for x in range(numdays)]
print(f" min date {min(date_list)}  num dates {len(date_list)} ")
for dt in date_list:
    year = dt.year
    month2 = make2(dt.month)
    day2 = make2(dt.day)
    daystr = dt.strftime("%Y-%m-%d")
    url = f"https://www.ncaa.com/scoreboard/basketball-men/d1/{year}/{month2}/{day2}/all-conf"
    try:
        response = requests.get(url)
        pass
    except Exception:
        msg = f"Error getting json from url {url} {datetime.datetime.now()}"
        msg += str(sys.exc_info())
        raise RuntimeError(msg)    
    soup = BeautifulSoup(response.text, "html.parser")
    elems = soup.findAll('div', {'class': ['status-final']})
    lsts = soup.findAll('ul', {'class': ['gamePod-game-teams']})
    #print(len(lsts))
    names = []
    scores = []
    games = []
    for i, lst in enumerate(lsts):
        #print(lst)
        z = soup.findAll('span', {'class':"gamePod-game-team-score"}, text=lst)
        temp = lst.findChildren('span', {'class': "gamePod-game-team-name"}, recursive=True, )
        names = [t.text for t in temp]
        temp = lst.findChildren('span', {'class': "gamePod-game-team-score"}, recursive=True)
        scores = [t.text for t in temp]
        games += [names + scores]
    df = pd.DataFrame(data=games, columns=["Team1", "Team2", "Score1", "Score2"])
    df["date"] = daystr
    print(f"{daystr} {df.shape[0]}")
    dfs.append(df)
rdf = pd.concat(dfs)
display(rdf["date"].value_counts())
print(rdf.shape)
```
